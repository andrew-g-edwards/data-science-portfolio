```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Desktop/data portfolio resources/life_data_week.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
hm = df.copy()
hm.drop('St.Dev', axis=1, inplace=True)
hm.drop('Unnamed: 14', axis=1, inplace=True)
hm.drop('week number', axis=1, inplace=True)
hm.head()
```

```python
hm.hist(bins = 20, color = 'green', figsize = (19, 15))
```

```python
hm = hm.corr(method='pearson')
plt.figure(figsize=(14,5))
map = sns.heatmap(hm, annot = True, fmt = '.1g', linewidths = .5, linecolor = 'Black',cmap='plasma', center=0, vmin= -.8, vmax=1)
map.set_title("Correlation of WEEKLY Behavior")
map.set_xticklabels(map.get_xticklabels(), rotation=45)
```

```python
day_df = pd.read_csv('Desktop/data portfolio resources/life_data_day.csv')
hmd = day_df.copy()
hmd.drop('weather', axis=1, inplace=True)
hmd.drop('shower', axis=1, inplace=True)
hmd.drop('day of the week', axis=1, inplace=True)
hmd.drop('month number', axis=1, inplace=True)
hmd.drop('week number', axis=1, inplace=True)
hmd.drop('day of the week number', axis=1, inplace=True)
hmd.drop('day', axis=1, inplace=True)
hmd.head()
```

```python
hmd = hmd.corr(method='pearson')
plt.figure(figsize=(14,5))
map = sns.heatmap(hmd, annot = True, fmt = '.1g', linewidths = .5, linecolor = 'Black',
                  cmap='viridis', center=0, vmin= -.8, vmax=1)
map.set_title("Correlation of DAILY Behavior")
map.set_xticklabels(map.get_xticklabels(), rotation=45)
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df, y='mood', x='socializing ', color='plum', 
            line_kws={"color":"red"}).set(title='Mood vs. Socializing, by week',ylabel='Average Weekly Mood',
                                         xlabel='Average Weekly Socialness')
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df, y='screen time (hrs)', x='outside (hrs)', color='lightgreen', 
            line_kws={"color":"green"}).set(title='Screen Time vs. Outside Time, by week'
                                           ,xlabel='Average Daily Hours Outside',
                                            ylabel='Average Daily Hours at Screen')
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df, y='food', x='excercise', color='cyan', 
            line_kws={"color":"blue"}).set(title='Appetite vs. Exercise, by week',
                                           xlabel='exercise', ylabel='food consumption')
```
