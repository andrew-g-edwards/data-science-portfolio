```python
chris = df.query('christian == 1').reset_index()
andy = df.query('andy == 1').reset_index()
bob = df.query('bob == 1').reset_index()
```

```python
chris.hist(bins = 20, color = 'green', figsize = (20, 9), 
           column = ['duration_mins','popularity','energy','tempo','valence',
                    'danceability','year'], layout=(2,4))
print("Christian's top 100 songs, distribution of attributes")
```

```python
bob.hist(bins = 20, color = 'red', figsize = (20, 9), 
           column = ['duration_mins','popularity','energy','tempo','valence',
                    'danceability','year'], layout=(2,4))
print("Bobby's top 100 songs, distribution of attributes")
```

```python
andy.hist(bins = 20, color = 'grey', figsize = (20, 9), 
           column = ['duration_mins','popularity','energy','tempo','valence',
                    'danceability','year'], layout=(2,4))
print("Andrew's top 100 songs, distribution of attributes")
```

```python
boys_tempo = pd.DataFrame()
boys_tempo['andy_tempo'] = andy['tempo']
boys_tempo['bob_tempo'] = bob['tempo']
boys_tempo['chris_tempo'] = chris['tempo']
```

```python
sns.kdeplot(boys_tempo, palette = 'bright',fill=True)
print("Distribution of Song Tempo")
```

```python
boys_year = pd.DataFrame()
boys_year['andy_year'] = andy['year']
boys_year['bob_year'] = bob['year']
boys_year['chris_year'] = chris['year']
sns.kdeplot(boys_year, palette = 'bright',fill=True)
print("Distribution of Release Date Year")
```

```python
boys_energy = pd.DataFrame()
boys_energy['andy_energy'] = andy['energy']
boys_energy['bob_energy'] = bob['energy']
boys_energy['chris_energy'] = chris['energy']
sns.kdeplot(boys_energy, palette = 'bright',fill=True)
print("Distribution of Song's Energy")
```

```python
boys_danceability = pd.DataFrame()
boys_danceability['andy_danceability'] = andy['danceability']
boys_danceability['bob_danceability'] = bob['danceability']
boys_danceability['chris_danceability'] = chris['danceability']
sns.kdeplot(boys_danceability, palette = 'bright',fill=True)
print("Distribution of Track's Danceability")
```
